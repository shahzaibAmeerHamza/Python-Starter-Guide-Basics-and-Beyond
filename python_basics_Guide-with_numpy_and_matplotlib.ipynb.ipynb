**Loops: Repetition of Code**

```python
# A simple for loop using range()
for i in range(5):
    print("This is loop number:", i)
```

```python
# A list of fruits
fruits = ["apple", "banana", "cherry"]

# For loop to iterate through the list
for fruit in fruits:
    print("I like", fruit)
```

```python
# Initializing a variable
x = 0

# While loop to repeat until x is no longer less than 5
while x < 5:
    print("The value of x is:", x)
    x += 1
```

```python
# For loop with break and continue
for i in range(10):
    if i == 5:
        break  # Exit the loop when i equals 5
    if i % 2 == 0:
        continue  # Skip the rest of the loop if i is even
    print(i)
```

```python
# Nested loop example
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i = {i}, j = {j}")
```

**Functions, Modularity of Code**

```python
# Define a function called greet
def greet():
    print("Hello, welcome to Python programming!")

# Call the function
greet()
```

```python
# Function that takes a name as an argument
def greet_user(name):
    print(f"Hello, {name}! Welcome to Python programming!")

# Call the function with your name
greet_user("Shahzaib Hamza")
```

```python
# Function to add two numbers
def add_numbers(a, b):
    return a + b

# Call the function and store the result
result = add_numbers(5, 3)
print("The sum is:", result)
```

```python
# Function with a default argument
def greet(name, message="Welcome!"):
    print(f"Hello, {name}! {message}")

# Call the function with and without the second argument
greet("Ali")
greet("Sana", "How are you today?")
```

```python
# Function to calculate the area of a circle
def area_of_circle(radius):
    pi = 3.14159
    return pi * radius * radius

# Use the function with different radii
print("Area with radius 5:", area_of_circle(5))
print("Area with radius 10:", area_of_circle(10))
```

**Working with Lists and Tuples**

```python
# Create a list of your favorite fruits
favorite_fruits = ["apple", "banana", "mango", "grape"]

# Print the list
print("My favorite fruits are:", favorite_fruits)
```

```python
#printing datatype of the list
print(type(favorite_fruits))
```

```python
# Access the first and last element
print("First fruit:", favorite_fruits[0])
print("Last fruit:", favorite_fruits[-1])

# Slice the list to get the first two fruits
print("First two fruits:", favorite_fruits[0:2])
```

```python
# Add a new fruit to the list
favorite_fruits.append("orange")
print("Updated list after adding orange:", favorite_fruits)

# Remove an item from the list
favorite_fruits.remove("banana")
print("Updated list after removing banana:", favorite_fruits)

# Change an element in the list
favorite_fruits[1] = "strawberry"
print("Updated list after changing second fruit:", favorite_fruits)
```

```python
# Use len() to find the length of the list
print("Number of fruits:", len(favorite_fruits))

# Use .sort() to sort the list alphabetically
favorite_fruits.sort()
print("Alphabetically sorted fruits:", favorite_fruits)

# Use .reverse() to reverse the list order
favorite_fruits.reverse()
print("Reversed list:", favorite_fruits)
```

**Creating and acessesing tuples**

```python
# Create a tuple of colors
colors = ("red", "green", "blue", "yellow")

# Access elements in the tuple
print("First color:", colors[0])
print("Last color:", colors[-1])
```

```python
# Trying to change an element in the tuple (this will cause an error)
try:
    colors[0] = "purple"
except TypeError as e:
    print("Error:", e)

# Converting a tuple to a list
colors_list = list(colors)
colors_list.append("purple")
print("Converted and modified list:", colors_list)
```

**Dictionaries in Python**

```python
# Create a dictionary with student details
student = {
    "name": "Shahzaib Hamza",
    "age": 22,
    "major": "Software Engineering"
}

# Print the entire dictionary
print("Student details:", student)

# Print individual values using keys
print("Name:", student["name"])
print("Age:", student["age"])
print("Major:", student["major"])
```

```python
# Add a new key-value pair
student["grade"] = "A"
print("Updated student details:", student)

# Update an existing value
student["age"] = 21
print("Updated age:", student)
```

```python
# Remove a key-value pair using del
del student["major"]
print("After removing major:", student)

# Using .pop() method to remove an item
grade = student.pop("grade")
print("Removed grade:", grade)
print("After removing grade:", student)
```

```python
# Loop through keys
for key in student:
    print("Key:", key)

# Loop through values
for value in student.values():
    print("Value:", value)

# Loop through both keys and values
for key, value in student.items():
    print(f"{key}: {value}")
```

**Error and Exception Handling in Python**


Exercise 1: Understanding Try-Except Blocks


```python
# Demonstrating exception handling
try:
    result = 10 / 0  # Intentional error
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

Exercise 2: Catching Multiple Exceptions

```python
try:
    number = int(input("Enter a number: "))
    result = 100 / number
    print("Result:", result)
except ZeroDivisionError:
    print("Error: You cannot divide by zero.")
except ValueError:
    print("Error: Please enter a valid number.")

```

Exercise 3: Else Block with Try-Except


```python
try:
    age = int(input("Enter your age: "))
except ValueError:
    print("Error: Please enter a valid number.")
else:
    print("Thank you! Your age is:", age)
```

Exercise 4: Ensuring Cleanup with Finally


```python
try:
    file = open("example.txt", "r")
    content = file.read()
    print("File content:", content)
except FileNotFoundError:
    print("Error: File does not exist.")
finally:
    print("End of the file operation.")
```

Exercise 5: Raising an Exception


```python
# Raising a custom exception
def check_number(num):
    if num < 0:
        raise ValueError("Error: Negative numbers are not allowed.")
    print("You entered:", num)

try:
    check_number(-5)  # Intentional error
except ValueError as e:
    print(e)
```
